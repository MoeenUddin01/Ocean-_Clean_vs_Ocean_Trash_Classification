# Balance Dataset to 7,200 Images Per Class

This notebook randomly samples 7,200 images from each class and saves them to a balanced dataset.

```python
import os
import random
import shutil
from pathlib import Path

# Set random seed for reproducibility
random.seed(42)

# Paths
source_path = Path('/home/moeenuddin/Desktop/Deep_learning/ocean/Ocean-_Clean_vs_Ocean_Trash_Classification/dataset/raw')
balance_path = source_path / 'balance_data'

# Target count per class
TARGET_COUNT = 7200

# Image extensions
image_extensions = {'.jpg', '.jpeg', '.png', '.gif', '.bmp', '.tiff', '.webp'}

print(f"Source: {source_path}")
print(f"Target: {balance_path}")
print(f"Target count per class: {TARGET_COUNT}")
```

```python
# Create balance_data directory
if balance_path.exists():
    print(f"Removing existing {balance_path}")
    shutil.rmtree(balance_path)

balance_path.mkdir(parents=True, exist_ok=True)

# Process each class
class_names = ['ocean_clean', 'ocean_trash']
results = {}

for class_name in class_names:
    class_source = source_path / class_name
    class_target = balance_path / class_name
    
    if not class_source.exists():
        print(f"⚠️  Class directory not found: {class_source}")
        continue
    
    # Get all image files
    image_files = [f for f in class_source.iterdir() 
                   if f.is_file() and f.suffix.lower() in image_extensions]
    
    available = len(image_files)
    print(f"\n📁 {class_name}:")
    print(f"   Available: {available:,} images")
    
    if available < TARGET_COUNT:
        print(f"   ⚠️  Not enough images! Need {TARGET_COUNT}, have {available}")
        selected = image_files
        results[class_name] = len(selected)
    else:
        # Randomly sample
        selected = random.sample(image_files, TARGET_COUNT)
        print(f"   ✓ Selected: {len(selected):,} images")
        results[class_name] = len(selected)
    
    # Create target directory
    class_target.mkdir(parents=True, exist_ok=True)
    
    # Copy selected images
    copied = 0
    for img_path in selected:
        target_file = class_target / img_path.name
        shutil.copy2(img_path, target_file)
        copied += 1
    
    print(f"   ✓ Copied: {copied:,} images")

print("\n" + "="*50)
```

```python
# Verify the balanced dataset
print("📊 VERIFICATION - Balanced Dataset Counts:")
print("="*40)

total = 0
for class_name in sorted(balance_path.iterdir()):
    if class_name.is_dir():
        count = len([f for f in class_name.iterdir() if f.is_file()])
        print(f"  {class_name.name}: {count:,} images")
        total += count

print("-"*40)
print(f"  TOTAL: {total:,} images")
print("="*40)
```
