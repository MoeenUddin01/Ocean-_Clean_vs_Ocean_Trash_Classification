# Image Count by Class

This notebook counts the number of images in each class of the dataset.

```python
import os
from pathlib import Path

# Define the dataset path
dataset_path = Path('/home/moeenuddin/Desktop/Deep_learning/ocean/Ocean-_Clean_vs_Ocean_Trash_Classification/dataset/raw')

# Image extensions to look for
image_extensions = {'.jpg', '.jpeg', '.png', '.gif', '.bmp', '.tiff', '.webp'}

# Dictionary to store counts
class_counts = {}

# Iterate through each class directory
for class_dir in sorted(dataset_path.iterdir()):
    if class_dir.is_dir():
        class_name = class_dir.name
        
        # Count image files
        image_count = 0
        for file_path in class_dir.iterdir():
            if file_path.is_file() and file_path.suffix.lower() in image_extensions:
                image_count += 1
        
        class_counts[class_name] = image_count

# Display results
print("=" * 40)
print("📊 IMAGE COUNT BY CLASS")
print("=" * 40)
total = 0
for class_name, count in sorted(class_counts.items()):
    print(f"  {class_name}: {count:,} images")
    total += count
print("-" * 40)
print(f"  TOTAL: {total:,} images")
print("=" * 40)
```

```python
# Optional: Show detailed breakdown with file extension distribution
print("\n📁 Detailed Breakdown by File Extension:\n")

for class_dir in sorted(dataset_path.iterdir()):
    if class_dir.is_dir():
        class_name = class_dir.name
        extension_counts = {}
        
        for file_path in class_dir.iterdir():
            if file_path.is_file():
                ext = file_path.suffix.lower()
                extension_counts[ext] = extension_counts.get(ext, 0) + 1
        
        print(f"\n{class_name}:")
        for ext, count in sorted(extension_counts.items()):
            print(f"  {ext}: {count:,}")
```
